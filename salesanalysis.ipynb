```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data creation
data = {
    "Date": [
        "2024-01-01", "2024-01-02", "2024-01-03", "2024-01-04", "2024-01-05",
        "2024-01-01", "2024-01-02", "2024-01-03", "2024-01-04", "2024-01-05"
    ],
    "Store": ["A", "A", "A", "A", "A", "B", "B", "B", "B", "B"],
    "Category": [
        "Electronics", "Electronics", "Clothing", "Clothing", "Grocery",
        "Electronics", "Clothing", "Clothing", "Grocery", "Grocery"
    ],
    "Sales": [
        1500, 2000, 1000, 1200, 800,
        1800, 1100, 1300, 700, 900
    ],
    "Profit": [
        300, 400, 200, 240, 160,
        360, 220, 260, 140, 180
    ]
}
```

```python
# Create a DataFrame
df = pd.DataFrame(data)

# Convert 'Date' column to datetime format
df["Date"] = pd.to_datetime(df["Date"])

# 1. Total Sales by Store
sales_by_store = df.groupby("Store")["Sales"].sum()

# 2. Total Profit by Category
profit_by_category = df.groupby("Category")["Profit"].sum()

# 3. Daily Sales Trend
daily_sales = df.groupby("Date")["Sales"].sum()

# 4. Top Categories by Sales
top_categories = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

# Visualization
plt.figure(figsize=(14, 8))
```

```python
# Bar plot: Total Sales by Store
plt.subplot(2, 2, 1)
sales_by_store.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Total Sales by Store")
plt.xlabel("Store")
plt.ylabel("Total Sales")
```

```python

# Bar plot: Total Profit by Category
plt.subplot(2, 2, 2)
profit_by_category.plot(kind="bar", color="lightgreen", edgecolor="black")
plt.title("Total Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
```

```python
# Line plot: Daily Sales Trend
plt.subplot(2, 2, 3)
daily_sales.plot(kind="line", marker="o", color="coral")
plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")
```

```python
# Bar plot: Top Categories by Sales
plt.subplot(2, 2, 4)
top_categories.plot(kind="bar", color="lightblue", edgecolor="black")
plt.title("Top Categories by Sales")
plt.xlabel("Category")
plt.ylabel("Total Sales")
```

```python
plt.tight_layout()
plt.show()
```

```python
# Summary statistics
print("Sales by Store:\n", sales_by_store)
print("\nProfit by Category:\n", profit_by_category)
print("\nDaily Sales Trend:\n", daily_sales)
print("\nTop Categories by Sales:\n", top_categories)
```
